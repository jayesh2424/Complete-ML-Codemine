# Simple Linear Regression

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import Dataset
data = pd.read_csv('/content/salary_experience_data.csv')
```

```python
# info
data.info()
```

```python
# stats values
data.describe()
```

```python
data.head()
```

X -> independent value

-----------------------
y -> dependent value (Target Value)

iloc[row, column]

------------------------------------------


iloc[:] -> all rows

----------------------------------------------

iloc[:-1]-> all columns except last (last is -1)

```python
# independent value (features i.e column which will decide my target value (salary))
X = data.iloc[:, :-1].values
y = data.iloc[:,-1].values
```

## Train - Test Split
 Here we are seperating (dividing the data ) into train data and test data.

 X -> X_train, X_test

 --------------------------------------------
 y -> y_train, y_test

```python
# import libraries
from sklearn.model_selection import train_test_split
```

random_state is our shuffling parameter.
if data is not sorted then no need of random state
but, in ideal situation you should always give the random state.

```python
# splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

# Train the Model

## Simple Linear Regression

```python
# import libraries
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# test the results
# y is target value
y_predict = regressor.predict(X_test)
y_predict
```

# Plot the Results

```python
# visualize the results
plt.scatter(X_train, y_train, color = 'red') # actual values from data
plt.plot(X_train, regressor.predict(X_train), color = 'blue') # predicted values
plt.title('Salary Prediction')
plt.xlabel('Years of Experience') # independent value
plt.ylabel('Salary') # dependent value
plt.show()
```

# Calculate Accuracy

```python
# libraries
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_test, y_predict)
print(f"R-Squared (R2): {r2: .2f}")
```
